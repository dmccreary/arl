Sample DonkeyCar Directories using GraphVis

For shape attributes are this site: https://www.graphviz.org/doc/info/attrs.html

For shapes types see: https://www.graphviz.org/doc/info/shapes.html#polygon

```python
from graphviz import Digraph

dot = Digraph(comment='DonkeyCar Directories')

dot.node('H', '/home/donkeycar', color="blue", style='filled', fontcolor="white")
dot.node('M', 'mycar')
dot.node('D', 'data')
dot.node('T', 'tub_1_DATE')
dot.node('J', 'X.json')
dot.node('I', 'X.jpg')
dot.node('U', 'tub_2_DATE')
dot.node('O', 'models')
dot.node('E', 'mymodel.h5', shape="box", height=".001", style='filled', color="lightgray")
dot.node('P', 'projects')
dot.node('N', 'donkeycar')
dot.node('C', 'config.py', shape="box", height=".001", style='filled', color="lightgray")
dot.node('Y', 'myconfig.py', shape="box", height=".001", style='filled', color="lightgray")


dot.edges(['HM', 'HP', 'MD', 'DT', 'DU', 'MO', 'MC', 'MY', 'PN', 'OE', 'TJ', 'TI'])
# the constraint puts the link on the same line
dot.edge('J', 'I', constraint='false')

dot
```
